```python
# Data mapping and pHash duplicate detection (exact hash matches first) - robust PIL + EXIF
import os, json, time
import pandas as pd
from PIL import Image, ImageOps, ImageFile
import imagehash
from tqdm import tqdm

# Robust PIL settings
Image.MAX_IMAGE_PIXELS = None
ImageFile.LOAD_TRUNCATED_IMAGES = True

t0=time.time()
train_csv = 'train.csv'
assert os.path.exists(train_csv), 'train.csv missing'
df = pd.read_csv(train_csv)
df['image'] = df['image'].astype(str)
df['chain'] = df['chain'].astype(str)
df['hotel_id'] = df['hotel_id'].astype(int)
df['train_path'] = 'train_images/' + df['chain'] + '/' + df['image']
df_test = pd.read_csv('sample_submission.csv')[['image']].copy()
df_test['test_path'] = 'test_images/' + df_test['image'].astype(str)

print('Train rows:', len(df), 'Unique hotels:', df['hotel_id'].nunique(), 'Chains:', df['chain'].nunique(), flush=True)
print('Test rows:', len(df_test), flush=True)

def phash_image(path):
    try:
        with Image.open(path) as im:
            im = ImageOps.exif_transpose(im.convert('RGB'))
            return str(imagehash.phash(im))
    except Exception as e:
        return None

# Compute pHash for train
train_phash_path = 'train_phash.csv'
if os.path.exists(train_phash_path):
    df_train_hash = pd.read_csv(train_phash_path)
else:
    paths = df['train_path'].tolist()
    hashes = []
    exists = 0
    for p in tqdm(paths, desc='phash train'):
        h = phash_image(p)
        hashes.append(h)
        if h is not None:
            exists += 1
    df_train_hash = df[['image','chain','hotel_id','train_path']].copy()
    df_train_hash['phash'] = hashes
    df_train_hash.to_csv(train_phash_path, index=False)
    print('Saved', train_phash_path, 'valid hashes:', exists, flush=True)

# Compute pHash for test
test_phash_path = 'test_phash.csv'
if os.path.exists(test_phash_path):
    df_test_hash = pd.read_csv(test_phash_path)
else:
    paths = df_test['test_path'].tolist()
    hashes = []
    exists = 0
    for p in tqdm(paths, desc='phash test'):
        h = phash_image(p)
        hashes.append(h)
        if h is not None:
            exists += 1
    df_test_hash = df_test.copy()
    df_test_hash['phash'] = hashes
    df_test_hash.to_csv(test_phash_path, index=False)
    print('Saved', test_phash_path, 'valid hashes:', exists, flush=True)

# Build exact-duplicate matches: test phash matching any train phash
train_hash_map = df_train_hash.dropna(subset=['phash']).groupby('phash').agg(list).reset_index()
train_hash_to_hotel = {row['phash']: row['hotel_id'] for _, row in train_hash_map.iterrows()}
train_hash_to_img = {row['phash']: row['image'] for _, row in train_hash_map.iterrows()}

matches = []
for _, r in df_test_hash.iterrows():
    h = r.get('phash', None)
    if pd.notna(h) and h in train_hash_to_hotel:
        hotels = train_hash_to_hotel[h]
        imgs = train_hash_to_img[h]
        # choose majority hotel_id among duplicates; fallback to first
        hid = pd.Series(hotels).mode().iloc[0]
        mt_img = imgs[0] if isinstance(imgs, list) and len(imgs)>0 else None
        matches.append({'image': r['image'], 'hotel_id': int(hid), 'match_type': 'exact_phash', 'matched_train_image': mt_img})

df_matches = pd.DataFrame(matches)
dup_map_path = 'phash_exact_matches.csv'
df_matches.to_csv(dup_map_path, index=False)
print('Exact pHash matches:', len(df_matches), 'saved to', dup_map_path, flush=True)
print('Done pHash in', round(time.time()-t0,1), 's', flush=True)
```

```python
# CLIP embeddings (ViT-L/14@336) for train/test with AMP and caching
import os, time, math, gc
import numpy as np
import pandas as pd
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import open_clip
from tqdm import tqdm

t0 = time.time()
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Load train/test mappings if not already in memory
if 'df' not in globals():
    df = pd.read_csv('train.csv')
    df['image'] = df['image'].astype(str)
    df['chain'] = df['chain'].astype(str)
    df['hotel_id'] = df['hotel_id'].astype(int)
    df['train_path'] = 'train_images/' + df['chain'] + '/' + df['image']
if 'df_test' not in globals():
    df_test = pd.read_csv('sample_submission.csv')[['image']].copy()
    df_test['test_path'] = 'test_images/' + df_test['image'].astype(str)

print('Train images:', len(df), 'Test images:', len(df_test), flush=True)

class ImgDataset(Dataset):
    def __init__(self, paths, transform):
        self.paths = paths
        self.transform = transform
    def __len__(self):
        return len(self.paths)
    def __getitem__(self, idx):
        p = self.paths[idx]
        with Image.open(p) as im:
            im = im.convert('RGB')
            im = self.transform(im)
        return im

def try_batch_sizes(candidates):
    for bs in candidates:
        try:
            x = torch.zeros(bs, 3, 336, 336, device=device, dtype=torch.float16)
            del x; torch.cuda.empty_cache()
            return bs
        except Exception:
            torch.cuda.empty_cache()
            continue
    return 8

def compute_embeddings(model, transform, paths, out_path, num_workers=2):
    if os.path.exists(out_path):
        arr = np.load(out_path)
        print('Loaded cached', out_path, arr.shape, flush=True)
        return arr
    ds = ImgDataset(paths, transform)
    bs = try_batch_sizes([128, 64, 48, 32, 24, 16, 12, 8])
    print(f'Using batch size {bs} on {device}', flush=True)
    dl = DataLoader(ds, batch_size=bs, shuffle=False, num_workers=num_workers, pin_memory=True)
    embs = []
    model.eval()
    seen = 0
    t_last = time.time()
    with torch.no_grad(), torch.cuda.amp.autocast(enabled=True):
        for i, batch in enumerate(dl):
            batch = batch.to(device, non_blocking=True)
            feats = model.encode_image(batch)
            feats = F.normalize(feats.float(), dim=-1)
            embs.append(feats.cpu().numpy())
            seen += batch.size(0)
            if (i+1) % 20 == 0 or seen == len(ds):
                dt = time.time()-t_last; t_last=time.time()
                print(f'  Encoded {seen}/{len(ds)} imgs; last20 iters dt={dt:.1f}s', flush=True)
    arr = np.concatenate(embs, axis=0)
    np.save(out_path, arr)
    print('Saved', out_path, arr.shape, flush=True)
    del embs; gc.collect(); torch.cuda.empty_cache()
    return arr

# Create model and preprocess
print('Loading CLIP ViT-L/14@336...', flush=True)
model, _, preprocess = open_clip.create_model_and_transforms('ViT-L-14-336', pretrained='openai', device=device)
print('Model loaded.', flush=True)

# Ensure preprocess res is 336; open_clip provides correct transform
train_paths = df['train_path'].tolist()
test_paths = df_test['test_path'].tolist()

# Compute and cache embeddings
train_emb_path = 'train_clip_vitl14_336.npy'
test_emb_path = 'test_clip_vitl14_336.npy'

train_emb = compute_embeddings(model, preprocess, train_paths, train_emb_path)
test_emb = compute_embeddings(model, preprocess, test_paths, test_emb_path)

# Save mapping CSVs for later retrieval
df[['image','chain','hotel_id','train_path']].to_csv('train_index.csv', index=False)
df_test[['image','test_path']].to_csv('test_index.csv', index=False)
print('Embeddings done in', round(time.time()-t0,1), 's', flush=True)
```

```python
# FAISS retrieval with TTA, soft chain gating, pHash override/re-ranking, one-step AQE, submission generation
import os, time, math
import numpy as np
import pandas as pd
import faiss
from tqdm import tqdm

t0 = time.time()

# Load indices and embeddings
train_index_csv = 'train_index.csv'
test_index_csv = 'test_index.csv'
train_emb_path = 'train_clip_vitl14_336.npy'
test_emb_path = 'test_clip_vitl14_336.npy'
assert os.path.exists(train_index_csv) and os.path.exists(test_index_csv), 'Index CSVs missing; run embedding cell first.'
assert os.path.exists(train_emb_path) and os.path.exists(test_emb_path), 'Embeddings missing; run embedding cell first.'
df_train_idx = pd.read_csv(train_index_csv)
df_test_idx = pd.read_csv(test_index_csv)
train_emb = np.load(train_emb_path).astype('float32')
test_emb = np.load(test_emb_path).astype('float32')

# Optional: test TTA average if available
tta_path = 'test_clip_vitl14_336_hflip.npy'
if os.path.exists(tta_path):
    test_emb_flip = np.load(tta_path).astype('float32')
    # average then renormalize per vector
    test_emb = test_emb + test_emb_flip
    norms = np.linalg.norm(test_emb, axis=1, keepdims=True) + 1e-12
    test_emb = test_emb / norms
    print('Applied test hflip TTA averaging:', test_emb.shape, flush=True)

# Normalize (safety) for cosine via inner product
train_emb = train_emb / (np.linalg.norm(train_emb, axis=1, keepdims=True) + 1e-12)
test_emb = test_emb / (np.linalg.norm(test_emb, axis=1, keepdims=True) + 1e-12)
print('Emb shapes:', train_emb.shape, test_emb.shape, flush=True)

# Prepare hotel_id and chain mapping
hotel_ids = df_train_idx['hotel_id'].astype(int).to_numpy()
# Factorize chains from strings to integer codes
chain_codes, chain_uni = pd.factorize(df_train_idx['chain'].astype(str))
chains = chain_codes.astype(np.int32)
unique_hotels, hotel_to_compact = np.unique(hotel_ids, return_inverse=True)
hotel_compact = hotel_to_compact.astype(np.int32)
num_hotels = unique_hotels.shape[0]
print('Unique hotels:', num_hotels, flush=True)

# Load pHash tables if present
train_phash_int = None; test_phash_int = None
if os.path.exists('train_phash.csv') and os.path.exists('test_phash.csv'):
    dtr = pd.read_csv('train_phash.csv')
    dte = pd.read_csv('test_phash.csv')
    # merge to align with index ordering
    dtr = df_train_idx.merge(dtr[['train_path','phash']], on='train_path', how='left')
    dte = df_test_idx.merge(dte[['test_path','phash']], on='test_path', how='left')
    def phash_hex_to_int(x):
        try:
            return int(str(x), 16)
        except Exception:
            return -1
    train_phash_int = dtr['phash'].apply(phash_hex_to_int).to_numpy(np.int64)
    test_phash_int = dte['phash'].apply(phash_hex_to_int).to_numpy(np.int64)
    print('Loaded pHash ints:', (train_phash_int!=-1).sum(), (test_phash_int!=-1).sum(), flush=True)

# Build FAISS index (IP for cosine on normalized vectors)
dim = train_emb.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(train_emb)
print('FAISS index ntotal:', index.ntotal, flush=True)

# Params
K = 200
batch = 1000
gamma = 3.0  # similarity exponent
rank_pow = 0.5  # rank weighting denom power
M_chain = 30   # top-M for chain majority
chain_ratio_thresh = 0.6
# AQE
aqe_k = 5
aqe_alpha = 0.3

def soft_chain_gate(weights, nbr_idx):
    if M_chain <= 0:
        return weights
    topm = nbr_idx[:min(M_chain, len(nbr_idx))]
    majority_chain = np.bincount(chains[topm]).argmax()
    r = (chains[topm] == majority_chain).mean()
    if r >= chain_ratio_thresh:
        mask = (chains[nbr_idx] == majority_chain)
        weights[~mask] *= 0.2  # soft gating
    return weights

def aggregate_scores(sim, idx, test_phash_int_slice):
    # sim: [N, K], idx: [N, K] over train indices
    N = sim.shape[0]
    preds_top5 = np.zeros((N, 5), dtype=np.int64)
    hid_compact_neighbors = hotel_compact[idx]
    # base weights: similarity^gamma with rank weighting
    ranks = np.arange(1, K+1, dtype=np.float32)[None, :]
    w = np.power(np.clip(sim, 0.0, 1.0), gamma) / np.power(ranks, rank_pow)

    for i in range(N):
        wi = w[i].copy()
        nbr_idx = idx[i]

        # Soft chain gating
        wi = soft_chain_gate(wi, nbr_idx)

        # pHash re-ranking boost within neighbors if available
        if train_phash_int is not None and test_phash_int_slice is not None:
            qh = int(test_phash_int_slice[i]) if test_phash_int_slice[i] is not None else -1
            th = train_phash_int[nbr_idx] if train_phash_int is not None else None
            d = np.full(len(nbr_idx), 64, dtype=np.int32)
            if qh != -1 and th is not None:
                valid = (th != -1)
                if np.any(valid):
                    x = np.bitwise_xor(th[valid].astype(np.int64), qh)
                    d_valid = np.fromiter((int(v).bit_count() for v in x), dtype=np.int32, count=int(valid.sum()))
                    d[valid] = d_valid
            exact_mask = (d == 0)
            if exact_mask.any():
                # force top1 to majority hotel among exact dup neighbors
                hids = hotel_ids[nbr_idx[exact_mask]]
                top_hid = pd.Series(hids).mode().iloc[0]
                preds_top5[i] = np.array([top_hid] * 5, dtype=np.int64)
                wi = None  # skip normal aggregation
            else:
                # d==1 strong tier, require at least 2
                d1 = (d == 1)
                if d1.sum() >= 2:
                    wi[d1] *= 3.5
                # 2<=d<=3 tier (exclude d==1), require at least 2
                close3 = (d <= 3) & (~d1)
                if close3.sum() >= 2:
                    wi[close3] *= 2.0
                # 4<=d<=5 light tier (exclude <=3)
                close5 = (d <= 5) & (~(d <= 3))
                wi[close5] *= 1.2

        if wi is None:
            continue

        scores = np.zeros(num_hotels, dtype=np.float32)
        np.add.at(scores, hid_compact_neighbors[i], wi)
        top5_c = np.argpartition(scores, -5)[-5:]
        top5_c = top5_c[np.argsort(scores[top5_c])[::-1]]
        preds_top5[i] = unique_hotels[top5_c]
    return preds_top5

def one_step_aqe(q_batch, sim, idx):
    # Build weighted neighbor mean per query with soft chain gating; return normalized updated queries
    N = q_batch.shape[0]
    ranks = np.arange(1, K+1, dtype=np.float32)[None, :]
    w = np.power(np.clip(sim, 0.0, 1.0), gamma) / np.power(ranks, rank_pow)  # [N,K]
    q_new = np.empty_like(q_batch)
    for i in range(N):
        wi = w[i].copy()
        nbr_idx = idx[i]
        wi = soft_chain_gate(wi, nbr_idx)
        # weighted mean of neighbor embeddings
        m = (train_emb[nbr_idx[:aqe_k]] * wi[:aqe_k, None]).sum(axis=0)
        ws = wi[:aqe_k].sum() + 1e-12
        m = m / ws
        q_prime = q_batch[i] + aqe_alpha * m
        q_prime = q_prime / (np.linalg.norm(q_prime) + 1e-12)
        q_new[i] = q_prime.astype('float32')
    return q_new

# Optional: pHash exact matches CSV (fast override map) for guaranteed top1 if exists
phash_override_map = {}
if os.path.exists('phash_exact_matches.csv'):
    try:
        dfm = pd.read_csv('phash_exact_matches.csv')
        for _, r in dfm.iterrows():
            ph = int(r['hotel_id'])
            phash_override_map[str(r['image'])] = ph
        print('Loaded exact pHash overrides:', len(phash_override_map), flush=True)
    except Exception as e:
        print('pHash override load failed:', e, flush=True)

# Global fallback: most frequent hotels in train
top_global = df_train_idx['hotel_id'].value_counts().index.to_numpy()
# Chain-aware backfill helpers
hid_to_chain = dict(zip(df_train_idx['hotel_id'].astype(int), df_train_idx['chain'].astype(str)))
chain_top = {c: g['hotel_id'].value_counts().index.to_numpy(dtype=int) for c, g in df_train_idx.groupby('chain')}

all_preds = []
N = test_emb.shape[0]
for s in range(0, N, batch):
    e = min(N, s+batch)
    q = test_emb[s:e]
    # Initial search
    sim1, idx1 = index.search(q, K)
    # AQE updated queries
    q_aqe = one_step_aqe(q, sim1, idx1)
    # Re-search with AQE
    sim2, idx2 = index.search(q_aqe, K)
    # Slice phash batch view
    test_phash_int_batch = test_phash_int[s:e] if test_phash_int is not None else None
    preds = aggregate_scores(sim2, idx2, test_phash_int_batch)
    all_preds.append(preds)
    print(f'Processed {e}/{N}', flush=True)
preds_all = np.vstack(all_preds)

# Apply exact pHash overrides to ensure top1 matches when available
test_images = df_test_idx['image'].astype(str).to_numpy()
for i, img in enumerate(test_images):
    if img in phash_override_map:
        hid = phash_override_map[img]
        cur = preds_all[i].tolist()
        out = [hid] + [x for x in cur if x != hid]
        if len(out) < 5:
            seen = set(out)
            for h in top_global:
                if len(out) >= 5: break
                if int(h) not in seen:
                    out.append(int(h))
        preds_all[i] = np.array(out[:5], dtype=np.int64)

# Ensure exactly 5 unique ids per row; chain-aware backfill first, then global
for i in range(len(preds_all)):
    row_list = list(dict.fromkeys(preds_all[i].tolist()))  # preserve order, unique
    seen = set(row_list)
    if len(row_list) < 5 and len(row_list) > 0:
        c = hid_to_chain.get(int(row_list[0]), None)
        if c is not None and c in chain_top:
            for h in chain_top[c]:
                if len(row_list) >= 5: break
                if int(h) not in seen:
                    row_list.append(int(h)); seen.add(int(h))
    if len(row_list) < 5:
        for h in top_global:
            if len(row_list) >= 5: break
            if int(h) not in seen:
                row_list.append(int(h)); seen.add(int(h))
    preds_all[i] = np.array(row_list[:5], dtype=np.int64)

# Build submission
sub = pd.DataFrame({'image': df_test_idx['image'].astype(str)})
sub['hotel_id'] = [' '.join(map(str, row)) for row in preds_all]
sub_path = 'submission.csv'
sub.to_csv(sub_path, index=False)
print('Wrote', sub_path, 'shape:', sub.shape, 'in', round(time.time()-t0,1), 's', flush=True)
```

```python
# Offline CV: 5-fold StratifiedKFold retrieval MAP@5 on train embeddings
import os, time
import numpy as np
import pandas as pd
import faiss
from sklearn.model_selection import StratifiedKFold

t0 = time.time()
train_index_csv = 'train_index.csv'
train_emb_path = 'train_clip_vitl14_336.npy'
assert os.path.exists(train_index_csv) and os.path.exists(train_emb_path), 'Run embedding cell first.'
df_train_idx = pd.read_csv(train_index_csv)
y = df_train_idx['hotel_id'].astype(int).to_numpy()
emb = np.load(train_emb_path).astype('float32')

# Normalize embeddings (safety)
emb = emb / (np.linalg.norm(emb, axis=1, keepdims=True) + 1e-12)

def map_at_5(y_true, preds_top5):
    # y_true: (N,), preds_top5: (N,5)
    hits = 0.0
    N = len(y_true)
    for i in range(N):
        row = preds_top5[i]
        for k in range(5):
            if row[k] == y_true[i]:
                hits += 1.0 / (k+1)
                break
    return hits / N

def fold_eval(train_idx, val_idx, K=200, gamma=3.0):
    xb = emb[train_idx]
    xq = emb[val_idx]
    ids_train = y[train_idx]
    ids_val = y[val_idx]
    # map hotel ids to compact indices
    uniq, inv = np.unique(ids_train, return_inverse=True)
    num_h = len(uniq)
    idx = faiss.IndexFlatIP(xb.shape[1])
    idx.add(xb)
    sim, nn = idx.search(xq, K)
    # aggregate scores by hotel
    hid_neighbors = inv[nn]  # map neighbor train indices -> compact hotel indices
    w = np.power(np.clip(sim, 0.0, 1.0), gamma)
    preds = np.zeros((len(val_idx), 5), dtype=np.int64)
    for i in range(xq.shape[0]):
        scores = np.zeros(num_h, dtype=np.float32)
        np.add.at(scores, hid_neighbors[i], w[i])
        top5_c = np.argpartition(scores, -5)[-5:]
        top5_c = top5_c[np.argsort(scores[top5_c])[::-1]]
        preds[i] = uniq[top5_c]
    score = map_at_5(ids_val, preds)
    return score

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
fold_scores = []
for f, (tr, va) in enumerate(skf.split(emb, y)):
    t1 = time.time()
    s = fold_eval(tr, va, K=200, gamma=3.0)
    fold_scores.append(s)
    print(f'Fold {f}: MAP@5={s:.5f} dt={time.time()-t1:.1f}s', flush=True)
cv = float(np.mean(fold_scores))
print('CV MAP@5:', f'{cv:.5f}', 'folds:', ' '.join(f'{x:.5f}' for x in fold_scores), 'elapsed', f'{time.time()-t0:.1f}s', flush=True)
```

```python
# Test hflip TTA embeddings (self-contained) with EXIF handling
import os, time, gc
import numpy as np
import pandas as pd
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image, ImageOps, ImageFile
import open_clip
from tqdm import tqdm

Image.MAX_IMAGE_PIXELS = None
ImageFile.LOAD_TRUNCATED_IMAGES = True

t0 = time.time()
device = 'cuda' if torch.cuda.is_available() else 'cpu'

df_test = pd.read_csv('test_index.csv') if os.path.exists('test_index.csv') else pd.read_csv('sample_submission.csv')[['image']].assign(test_path=lambda d: 'test_images/' + d['image'].astype(str))
paths = df_test['test_path'].tolist()
print('Test count:', len(paths), flush=True)

print('Loading CLIP ViT-L/14@336...', flush=True)
model, _, preprocess = open_clip.create_model_and_transforms('ViT-L-14-336', pretrained='openai', device=device)
model.eval()
print('Model loaded.', flush=True)

class TestFlipDS(Dataset):
    def __init__(self, paths, transform):
        self.paths = paths; self.transform = transform
    def __len__(self): return len(self.paths)
    def __getitem__(self, i):
        p = self.paths[i]
        with Image.open(p) as im:
            im = ImageOps.exif_transpose(im.convert('RGB'))
            im = im.transpose(Image.FLIP_LEFT_RIGHT)
            im = self.transform(im)
        return im

def try_bs(cands):
    for bs in cands:
        try:
            x = torch.zeros(bs, 3, 336, 336, device=device, dtype=torch.float16)
            del x; torch.cuda.empty_cache(); return bs
        except Exception:
            torch.cuda.empty_cache()
    return 8

out_path = 'test_clip_vitl14_336_hflip.npy'
if os.path.exists(out_path):
    arr = np.load(out_path)
    print('Loaded cached', out_path, arr.shape, flush=True)
else:
    ds = TestFlipDS(paths, preprocess)
    bs = try_bs([128, 64, 48, 32, 24, 16, 12, 8])
    print('Batch size:', bs, flush=True)
    dl = DataLoader(ds, batch_size=bs, shuffle=False, num_workers=2, pin_memory=True)
    embs = []
    seen = 0; t_last = time.time()
    with torch.no_grad(), torch.cuda.amp.autocast(enabled=True):
        for i, batch in enumerate(dl):
            batch = batch.to(device, non_blocking=True)
            feats = model.encode_image(batch)
            feats = F.normalize(feats.float(), dim=-1)
            embs.append(feats.cpu().numpy())
            seen += batch.size(0)
            if (i+1) % 20 == 0 or seen == len(ds):
                dt = time.time()-t_last; t_last=time.time()
                print(f'  Encoded {seen}/{len(ds)} test(hflip); last20 dt={dt:.1f}s', flush=True)
    arr = np.concatenate(embs, axis=0)
    np.save(out_path, arr)
    print('Saved', out_path, arr.shape, 'in', round(time.time()-t0,1), 's', flush=True)
    del embs; gc.collect(); torch.cuda.empty_cache()
```

```python
# Offline CV (mirrors inference): StratifiedGroupKFold with pHash groups, soft chain gating, rank-weighted agg, one-step AQE, pHash boosts/override
import os, time, numpy as np, pandas as pd, faiss
from sklearn.model_selection import StratifiedKFold, StratifiedGroupKFold

t0=time.time()
train_index_csv = 'train_index.csv'
train_emb_path = 'train_clip_vitl14_336.npy'
assert os.path.exists(train_index_csv) and os.path.exists(train_emb_path), 'Run embedding cell first.'
df_train_idx = pd.read_csv(train_index_csv)
y = df_train_idx['hotel_id'].astype(int).to_numpy()
# Factorize chain strings to integer codes (fix dtype bug)
chain_codes, chain_uni = pd.factorize(df_train_idx['chain'].astype(str))
chains = chain_codes.astype(np.int32)
emb = np.load(train_emb_path).astype('float32')
emb = emb / (np.linalg.norm(emb, axis=1, keepdims=True) + 1e-12)

# pHash groups for duplicate-safe CV if available + aligned pHash ints
groups = None
phash_int_full = None
train_phash_csv = 'train_phash.csv'
if os.path.exists(train_phash_csv):
    dtr_ph = pd.read_csv(train_phash_csv)[['train_path','phash']].copy()
    # Build mapping to avoid any merge explosion; first occurrence wins
    phash_map = dict(zip(dtr_ph['train_path'], dtr_ph['phash']))
    phash_col = df_train_idx['train_path'].map(phash_map)
    groups = phash_col.fillna('NA').astype(str).to_numpy()
    def phash_hex_to_int(x):
        try:
            return int(str(x), 16)
        except Exception:
            return -1
    phash_int_full = phash_col.apply(phash_hex_to_int).to_numpy(np.int64)
    assert len(groups) == len(df_train_idx), f'groups len {len(groups)} != train {len(df_train_idx)}'
    print('Using StratifiedGroupKFold with pHash groups (unique groups):', len(np.unique(groups)), flush=True)
else:
    print('train_phash.csv missing; falling back to StratifiedKFold', flush=True)

def map_at_5(y_true, preds_top5):
    hits = 0.0; N = len(y_true)
    for i in range(N):
        row = preds_top5[i]
        for k in range(5):
            if row[k] == y_true[i]:
                hits += 1.0 / (k+1); break
    return hits / N

# Params to mirror inference
K = 200; gamma = 3.0; rank_pow = 0.5; M_chain = 30; chain_ratio_thresh = 0.6
aqe_k = 5; aqe_alpha = 0.3

def soft_chain_gate(weights, nbr_abs_idx):
    if M_chain <= 0:
        return weights
    topm_abs = nbr_abs_idx[:min(M_chain, len(nbr_abs_idx))]
    majority_chain = np.bincount(chains[topm_abs]).argmax()
    r = (chains[topm_abs] == majority_chain).mean()
    if r >= chain_ratio_thresh:
        mask = (chains[nbr_abs_idx] == majority_chain)
        weights[~mask] *= 0.2
    return weights

def one_step_aqe(xq, xb, sim, nn_abs):
    # xq: [Nq,D], xb: [Nb,D], sim: [Nq,K], nn_abs: [Nq,K] absolute indices into full emb
    Nq = xq.shape[0]
    ranks = np.arange(1, K+1, dtype=np.float32)[None, :]
    w = np.power(np.clip(sim, 0.0, 1.0), gamma) / np.power(ranks, rank_pow)
    q_new = np.empty_like(xq)
    for i in range(Nq):
        wi = w[i].copy()
        nbr_abs = nn_abs[i]
        wi = soft_chain_gate(wi, nbr_abs)
        topk = min(aqe_k, len(nbr_abs))
        m = (emb[nbr_abs[:topk]] * wi[:topk, None]).sum(axis=0)
        ws = wi[:topk].sum() + 1e-12
        m = m / ws
        q_prime = xq[i] + aqe_alpha * m.astype(np.float32)
        q_prime = q_prime / (np.linalg.norm(q_prime) + 1e-12)
        q_new[i] = q_prime.astype('float32')
    return q_new

def fold_eval(train_idx, val_idx):
    xb = emb[train_idx]; xq = emb[val_idx]
    ids_train = y[train_idx]; ids_val = y[val_idx]
    uniq, inv = np.unique(ids_train, return_inverse=True)
    num_h = len(uniq)
    idx = faiss.IndexFlatIP(xb.shape[1])
    idx.add(xb)
    # initial search
    sim1, nn1_local = idx.search(xq, K)  # neighbors indices are local to xb
    # map to absolute indices to use chains from full set
    nn1_abs = train_idx[nn1_local]
    # AQE
    xq_aqe = one_step_aqe(xq, xb, sim1, nn1_abs)
    # re-search after AQE
    sim2, nn2_local = idx.search(xq_aqe, K)
    nn2_abs = train_idx[nn2_local]
    # rank weights
    ranks = np.arange(1, K+1, dtype=np.float32)[None, :]
    w = np.power(np.clip(sim2, 0.0, 1.0), gamma) / np.power(ranks, rank_pow)
    preds = np.zeros((len(val_idx), 5), dtype=np.int64)
    # map neighbors -> compact hotel indices
    hid_neighbors = inv[nn2_local]
    for i in range(xq.shape[0]):
        wi = w[i].copy()
        nbr_abs = nn2_abs[i]
        wi = soft_chain_gate(wi, nbr_abs)

        # pHash override/boost parity with inference, if available
        if phash_int_full is not None:
            qh = int(phash_int_full[val_idx[i]])
            th = phash_int_full[nbr_abs]
            d = np.full(len(nbr_abs), 64, dtype=np.int32)
            if qh != -1:
                valid = (th != -1)
                if np.any(valid):
                    x = np.bitwise_xor(th[valid].astype(np.int64), qh)
                    d_valid = np.fromiter((int(v).bit_count() for v in x), dtype=np.int32, count=int(valid.sum()))
                    d[valid] = d_valid
            exact_mask = (d == 0)
            if exact_mask.any():
                # force top1 to majority hotel among exact neighbors
                hids = ids_train[nn2_local[i][exact_mask]]
                top_hid = pd.Series(hids).mode().iloc[0]
                preds[i] = np.array([top_hid]*5, dtype=np.int64)
                continue  # skip normal aggregation
            else:
                # d==1 strong tier, require at least 2
                d1 = (d == 1)
                if d1.sum() >= 2:
                    wi[d1] *= 3.5
                # 2<=d<=3 tier (exclude d==1), require at least 2
                close3 = (d <= 3) & (~d1)
                if close3.sum() >= 2:
                    wi[close3] *= 2.0
                # 4<=d<=5 light tier (exclude <=3)
                close5 = (d <= 5) & (~(d <= 3))
                wi[close5] *= 1.2

        scores = np.zeros(num_h, dtype=np.float32)
        np.add.at(scores, hid_neighbors[i], wi)
        top5_c = np.argpartition(scores, -5)[-5:]
        top5_c = top5_c[np.argsort(scores[top5_c])[::-1]]
        preds[i] = uniq[top5_c]
    return map_at_5(ids_val, preds)

if groups is not None:
    # sanity
    assert len(groups) == len(emb) == len(y), f'Length mismatch groups:{len(groups)} emb:{len(emb)} y:{len(y)}'
    skf = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=42)
    splitter = skf.split(emb, y, groups)
else:
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    splitter = skf.split(emb, y)

fold_scores = []
for f, (tr, va) in enumerate(splitter):
    t1 = time.time()
    s = fold_eval(tr, va)
    fold_scores.append(s)
    print(f'Fold {f}: MAP@5={s:.5f} dt={time.time()-t1:.1f}s', flush=True)
cv = float(np.mean(fold_scores))
print('CV MAP@5:', f'{cv:.5f}', 'folds:', ' '.join(f'{x:.5f}' for x in fold_scores), 'elapsed', f'{time.time()-t0:.1f}s', flush=True)
```
