```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GroupKFold
from scipy.stats import pearsonr
import torch
from torch.utils.data import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset as HFDataset
import random
import importlib.metadata
random.seed(42); np.random.seed(42); torch.manual_seed(42)

# Load and subsample train data (10k rows for quick test)
train_df = pd.read_csv('train.csv')
subsample_idx = np.random.choice(len(train_df), 10000, replace=False)
train_df = train_df.iloc[subsample_idx].reset_index(drop=True)
print(f'Subsampled train shape: {train_df.shape}')

# Prepare inputs - return lists, not tensors
def prepare_features(examples):
    texts1 = [f"anchor: {anchor} [CPC] {context}" for anchor, context in zip(examples['anchor'], examples['context'])]
    texts2 = [f"target: {target} [CPC] {context}" for target, context in zip(examples['target'], examples['context'])]
    model_inputs = tokenizer(texts1, texts2, max_length=128, truncation=True, padding='max_length')
    model_inputs['labels'] = examples['score']
    return model_inputs

model_name = 'microsoft/deberta-v3-small'
tokenizer = AutoTokenizer.from_pretrained(model_name)

# GroupKFold by anchor, 3 splits for smoke test
gkf = GroupKFold(n_splits=3)
oof = np.zeros(len(train_df))
fold_pearsons = []

for fold, (train_idx, val_idx) in enumerate(gkf.split(train_df, groups=train_df['anchor'])):
    print(f'\n--- Fold {fold+1}/3 ---')
    train_fold = train_df.iloc[train_idx]
    val_fold = train_df.iloc[val_idx]
    
    # Create HF datasets
    train_dataset = HFDataset.from_pandas(train_fold[['anchor', 'target', 'context', 'score']])
    val_dataset = HFDataset.from_pandas(val_fold[['anchor', 'target', 'context', 'score']])
    
    # Tokenize
    train_dataset = train_dataset.map(prepare_features, batched=True, remove_columns=train_dataset.column_names)
    val_dataset = val_dataset.map(prepare_features, batched=True, remove_columns=val_dataset.column_names)
    
    # Model
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=1, problem_type='regression')
    
    # Patch importlib.metadata.version for torch to avoid PackageNotFoundError due to +cpu wheel
    original_version = importlib.metadata.version
    def patched_version(name):
        if name == "torch":
            return "2.4.1"
        return original_version(name)
    importlib.metadata.version = patched_version
    
    # Training args (CPU, fp32, 1 epoch, larger batch)
    training_args = TrainingArguments(
        output_dir=f'./fold_{fold}',
        num_train_epochs=1,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=32,
        learning_rate=2e-5,
        weight_decay=0.01,
        warmup_ratio=0.1,
        lr_scheduler_type='cosine',
        logging_steps=50,
        save_strategy='no',
        evaluation_strategy='no',
        fp16=False,
        dataloader_num_workers=0,
        report_to=None,
        load_best_model_at_end=False
    )
    
    # Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        tokenizer=tokenizer,
    )
    
    # Train
    trainer.train()
    
    # Predict OOF with symmetry TTA - fix swap
    def predict_tta(dataset):
        # Normal
        preds_normal = trainer.predict(dataset).predictions.flatten()
        # Swapped: create swapped DF
        swapped_df = val_fold.copy()
        swapped_df[['anchor','target']] = swapped_df[['target','anchor']]
        swapped = HFDataset.from_pandas(swapped_df[['anchor','target','context','score']])
        swapped = swapped.map(prepare_features, batched=True, remove_columns=swapped.column_names)
        preds_swapped = trainer.predict(swapped).predictions.flatten()
        return np.clip((preds_normal + preds_swapped) / 2, 0, 1)
    
    val_preds = predict_tta(val_dataset)
    oof[val_idx] = val_preds
    fold_pearson = pearsonr(val_preds, val_fold['score'].values)[0]
    fold_pearsons.append(fold_pearson)
    print(f'Fold {fold+1} Pearson: {fold_pearson:.4f}')

# Restore original version function (optional)
importlib.metadata.version = original_version

# Overall OOF
oof_pearson = pearsonr(oof, train_df['score'].values)[0]
print(f'\nSubsample OOF Pearson: {oof_pearson:.4f}')
print(f'Mean fold Pearson: {np.mean(fold_pearsons):.4f} +/- {np.std(fold_pearsons):.4f}')
if oof_pearson > 0.78:
    print('Viable smoke test! Proceed to full run or GPU attempt.')
else:
    print('Low score; further debug needed.')
```

```python
import pandas as pd, numpy as np
from sklearn.model_selection import GroupKFold
from scipy.stats import pearsonr
from sklearn.isotonic import IsotonicRegression
import torch, importlib.metadata
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding
from datasets import Dataset as HFDataset
import random, os
random.seed(42); np.random.seed(42); torch.manual_seed(42)

# Load data
train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')

# Try to use cpc_texts.csv if it exists; else fallback to enriched section map
use_titles = False
title_fallback = 'no title'
try:
    cpc_df = pd.read_csv('cpc_texts.csv')
    if 'context' not in cpc_df.columns and 'code' in cpc_df.columns:
        cpc_df = cpc_df.rename(columns={'code': 'context'})
    text_col = next((c for c in ['title','text','description'] if c in cpc_df.columns), None)
    if text_col is None or 'context' not in cpc_df.columns:
        raise FileNotFoundError
    cpc_df = cpc_df[['context', text_col]].drop_duplicates().rename(columns={text_col: 'title'})
    train_df = train_df.merge(cpc_df, on='context', how='left')
    test_df  = test_df.merge(cpc_df,  on='context', how='left')
    train_df['title'] = train_df['title'].fillna(title_fallback)
    test_df['title']  = test_df['title'].fillna(title_fallback)
    use_titles = True
except FileNotFoundError:
    sec_map = {
        'A': 'Human Necessities, Medical, Agriculture',
        'B': 'Operations, Transport, Separation, Mixing',
        'C': 'Chemistry, Metallurgy, Combinatorial Tech',
        'D': 'Textiles, Paper, Building Materials',
        'E': 'Fixed Constructions, Building, Earth Moving',
        'F': 'Mechanical Engineering, Lighting, Heating',
        'G': 'Physics, Computing, Calculating, Counting',
        'H': 'Electricity, Basic Electric Elements',
        'Y': 'Emerging Technologies, Cross-Sectional'
    }
    train_df['title'] = train_df['context'].str[0].map(sec_map).fillna(title_fallback)
    test_df['title']  = test_df['context'].str[0].map(sec_map).fillna(title_fallback)
    use_titles = False

print('Use titles:', use_titles, '| train:', train_df.shape, '| test:', test_df.shape)

model_name = 'microsoft/deberta-v3-small'
tokenizer = AutoTokenizer.from_pretrained(model_name)

def prepare_features(examples):
    if use_titles:
        t1 = [f"anchor: {a} [CPC] {c} {t}" for a,c,t in zip(examples['anchor'], examples['context'], examples['title'])]
        t2 = [f"target: {b} [CPC] {c} {t}" for b,c,t in zip(examples['target'], examples['context'], examples['title'])]
        mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    else:
        t1 = [f"anchor: {a} [CPC] {c}" for a,c in zip(examples['anchor'], examples['context'])]
        t2 = [f"target: {b} [CPC] {c}" for b,c in zip(examples['target'], examples['context'])]
        mi = tokenizer(t1, t2, max_length=128, truncation=True, padding=False)
    if 'score' in examples:
        mi['labels'] = examples['score']
    return mi

def prepare_test_features(examples):
    if use_titles:
        t1 = [f"anchor: {a} [CPC] {c} {t}" for a,c,t in zip(examples['anchor'], examples['context'], examples['title'])]
        t2 = [f"target: {b} [CPC] {c} {t}" for b,c,t in zip(examples['target'], examples['context'], examples['title'])]
        mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    else:
        t1 = [f"anchor: {a} [CPC] {c}" for a,c in zip(examples['anchor'], examples['context'])]
        t2 = [f"target: {b} [CPC] {c}" for b,c in zip(examples['target'], examples['context'])]
        mi = tokenizer(t1, t2, max_length=128, truncation=True, padding=False)
    return mi

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    preds = np.clip(preds.flatten(), 0, 1)
    return {'pearson': pearsonr(preds, labels)[0]}

gkf = GroupKFold(n_splits=5)
oof = np.zeros(len(train_df))
test_preds = np.zeros((5, len(test_df)))
fold_pearsons = []

# Patch torch metadata for +cpu wheel
original_version = importlib.metadata.version
def patched_version(name):
    if name == 'torch': return '2.4.1'
    return original_version(name)
importlib.metadata.version = patched_version

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

for fold, (tr_idx, va_idx) in enumerate(gkf.split(train_df, groups=train_df['anchor'])):
    print(f'\n--- Fold {fold+1}/5 ---')
    tr = train_df.iloc[tr_idx]
    va = train_df.iloc[va_idx]

    ds_tr = HFDataset.from_pandas(tr[['anchor','target','context','title','score']])
    ds_va = HFDataset.from_pandas(va[['anchor','target','context','title','score']])
    ds_te = HFDataset.from_pandas(test_df[['anchor','target','context','title']])

    ds_tr = ds_tr.map(prepare_features, batched=True, remove_columns=ds_tr.column_names)
    ds_va = ds_va.map(prepare_features, batched=True, remove_columns=ds_va.column_names)
    ds_te = ds_te.map(prepare_test_features, batched=True, remove_columns=ds_te.column_names)

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=1, problem_type='regression')

    args = TrainingArguments(
        output_dir=f'./fold_{fold}',
        num_train_epochs=3,
        per_device_train_batch_size=8,             # CPU-friendly
        per_device_eval_batch_size=16,
        gradient_accumulation_steps=2,             # effective 16
        learning_rate=2e-5,
        weight_decay=0.01,
        warmup_ratio=0.1,
        lr_scheduler_type='cosine',
        logging_steps=100,
        save_strategy='epoch',
        evaluation_strategy='epoch',
        load_best_model_at_end=True,
        metric_for_best_model='pearson',
        greater_is_better=True,
        save_total_limit=1,
        fp16=False,
        dataloader_num_workers=0,
        report_to=None,
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=ds_tr,
        eval_dataset=ds_va,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
    )
    trainer.train()

    def predict_tta(trainer, dataset, df, is_test=False):
        normal = trainer.predict(dataset).predictions.flatten()
        swapped_df = df.copy()
        swapped_df[['anchor','target']] = swapped_df[['target','anchor']]
        swapped = HFDataset.from_pandas(swapped_df[['anchor','target','context','title']] if is_test else swapped_df[['anchor','target','context','title','score']])
        swapped = swapped.map(prepare_test_features if is_test else prepare_features, batched=True, remove_columns=swapped.column_names)
        swapped_pred = trainer.predict(swapped).predictions.flatten()
        return np.clip((normal + swapped_pred)/2, 0, 1)

    va_pred = predict_tta(trainer, ds_va, va, is_test=False)
    oof[va_idx] = va_pred
    fold_pearson = pearsonr(va_pred, va['score'].values)[0]
    fold_pearsons.append(fold_pearson)
    print(f'Fold {fold+1} Pearson: {fold_pearson:.4f}')

    test_preds[fold] = predict_tta(trainer, ds_te, test_df, is_test=True)

importlib.metadata.version = original_version

oof_pearson = pearsonr(oof, train_df['score'].values)[0]
print(f'\nFull OOF Pearson: {oof_pearson:.4f}')
print(f'Mean fold Pearson: {np.mean(fold_pearsons):.4f} +/- {np.std(fold_pearsons):.4f}')

iso = IsotonicRegression(out_of_bounds='clip')
iso.fit(oof, train_df['score'].values)
oof_cal = iso.predict(oof)
oof_cal_pearson = pearsonr(oof_cal, train_df['score'].values)[0]
print(f'Calibrated OOF Pearson: {oof_cal_pearson:.4f}')

test_avg = np.mean(test_preds, axis=0)
test_final = np.clip(iso.predict(test_avg) if oof_cal_pearson > oof_pearson else test_avg, 0, 1)

submission = pd.DataFrame({'id': test_df['id'], 'score': test_final})
submission.to_csv('submission.csv', index=False)
print('Submission saved! Final OOF:', max(oof_pearson, oof_cal_pearson))
```

```python
import pandas as pd, numpy as np
from sklearn.model_selection import GroupKFold
from scipy.stats import pearsonr
from sklearn.isotonic import IsotonicRegression
import torch, importlib.metadata
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding, EarlyStoppingCallback
from datasets import Dataset as HFDataset
import random, os
random.seed(42); np.random.seed(42); torch.manual_seed(42)

# Load data
train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')

# Try to use cpc_texts.csv if it exists; else fallback to enriched section + class map
use_titles = False
title_fallback = 'no title'
try:
    cpc_df = pd.read_csv('cpc_texts.csv')
    if 'context' not in cpc_df.columns and 'code' in cpc_df.columns:
        cpc_df = cpc_df.rename(columns={'code': 'context'})
    text_col = next((c for c in ['title','text','description'] if c in cpc_df.columns), None)
    if text_col is None or 'context' not in cpc_df.columns:
        raise FileNotFoundError
    cpc_df = cpc_df[['context', text_col]].drop_duplicates().rename(columns={text_col: 'title'})
    train_df = train_df.merge(cpc_df, on='context', how='left')
    test_df  = test_df.merge(cpc_df,  on='context', how='left')
    train_df['title'] = train_df['title'].fillna(title_fallback)
    test_df['title']  = test_df['title'].fillna(title_fallback)
    use_titles = True
except FileNotFoundError:
    sec_map = {
        'A': 'Human Necessities, Medical, Agriculture',
        'B': 'Operations, Transport, Separation, Mixing',
        'C': 'Chemistry, Metallurgy, Combinatorial Tech',
        'D': 'Textiles, Paper, Building Materials',
        'E': 'Fixed Constructions, Building, Earth Moving',
        'F': 'Mechanical Engineering, Lighting, Heating',
        'G': 'Physics, Computing, Calculating, Counting',
        'H': 'Electricity, Basic Electric Elements',
        'Y': 'Emerging Technologies, Cross-Sectional'
    }
    class_map = {
        'A61': 'Medical or Veterinary Science; Hygiene',
        'C07': 'Organic Chemistry',
        'C08': 'Organic Macromolecular Compounds; Preparation or Chemical Working-Up',
        'C12': 'Biochemistry; Beer; Spirits; Wine; Vinegar; Microbiology; Enzymology; Mutation or Genetic Engineering',
        'G01': 'Measuring; Testing',
        'G06': 'Computing; Calculating; Counting',
        'H01': 'Basic Electric Elements',
        'H04': 'Electric Communication Technique',
        'B60': 'Vehicles in General',
        'F16': 'Engineering Elements or Units; General Measures for Producing and Maintaining Effective Functioning of Machines or Installations',
        'H05': 'Electric Techniques Not Otherwise Provided For',
        'H03': 'Basic Electronic Circuitry',
        'H01L': 'Semiconductor Devices; Electric Solid State Devices Not Otherwise Provided For',
        'A61K': 'Preparations for Medical, Dental, or Toilet Purposes'
    }
    def get_title(context):
        cls = context[:3] if len(context) >= 3 else context[0]
        sec = context[0]
        return class_map.get(cls, sec_map.get(sec, title_fallback))
    train_df['title'] = train_df['context'].apply(get_title)
    test_df['title']  = test_df['context'].apply(get_title)
    use_titles = True  # Use improved fallback

# Build hierarchical CPC tokens from train+test
all_ctx = pd.concat([train_df['context'], test_df['context']]).unique()

def get_hier_parts(code):
    sec = code[0] if code else 'X'
    cls = code[:3] if len(code) >= 3 else sec
    sub = code[:4] if len(code) >= 4 else cls
    grp = code[:7].replace('/', '') if len(code) >= 7 else sub  # e.g., A61K31
    return f"<SEC_{sec}>", f"<CLS_{cls}>", f"<SUB_{sub}>", f"<GRP_{grp}>"

special_tokens = set()
for c in all_ctx:
    special_tokens.update(get_hier_parts(c))

model_name = 'microsoft/deberta-v3-base'
tokenizer = AutoTokenizer.from_pretrained(model_name)
n_added = tokenizer.add_special_tokens({'additional_special_tokens': sorted(special_tokens)})
print(f"Added {n_added} hierarchical tokens. Vocab: {len(tokenizer)}")

def hier_str(code):
    return " ".join(get_hier_parts(code)) + " "

def prepare_features(examples):
    h = [hier_str(c) for c in examples['context']]
    if use_titles:
        t1 = [f"{hh}anchor: {a} [CPC] {c} {t}" for hh,a,c,t in zip(h, examples['anchor'], examples['context'], examples['title'])]
        t2 = [f"{hh}target: {b} [CPC] {c} {t}" for hh,b,c,t in zip(h, examples['target'], examples['context'], examples['title'])]
        mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    else:
        t1 = [f"{hh}anchor: {a} [CPC] {c}" for hh,a,c in zip(h, examples['anchor'], examples['context'])]
        t2 = [f"{hh}target: {b} [CPC] {c}" for hh,b,c in zip(h, examples['target'], examples['context'])]
        mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    if 'score' in examples:
        mi['labels'] = examples['score']
    return mi

def prepare_test_features(examples):
    h = [hier_str(c) for c in examples['context']]
    if use_titles:
        t1 = [f"{hh}anchor: {a} [CPC] {c} {t}" for hh,a,c,t in zip(h, examples['anchor'], examples['context'], examples['title'])]
        t2 = [f"{hh}target: {b} [CPC] {c} {t}" for hh,b,c,t in zip(h, examples['target'], examples['context'], examples['title'])]
        mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    else:
        t1 = [f"{hh}anchor: {a} [CPC] {c}" for hh,a,c in zip(h, examples['anchor'], examples['context'])]
        t2 = [f"{hh}target: {b} [CPC] {c}" for hh,b,c in zip(h, examples['target'], examples['context'])]
        mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    return mi

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    preds = np.clip(preds.flatten(), 0, 1)
    return {'pearson': pearsonr(preds, labels)[0]}

gkf = GroupKFold(n_splits=3)
oof_base = np.zeros(len(train_df))
test_preds_base = np.zeros((3, len(test_df)))
fold_pearsons_base = []

# Patch torch metadata for +cpu wheel
original_version = importlib.metadata.version
def patched_version(name):
    if name == 'torch': return '2.4.1'
    return original_version(name)
importlib.metadata.version = patched_version

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

for fold, (tr_idx, va_idx) in enumerate(gkf.split(train_df, groups=train_df['anchor'])):
    print(f'\n--- Fold {fold+1}/3 ---')
    tr = train_df.iloc[tr_idx]
    va = train_df.iloc[va_idx]

    ds_tr = HFDataset.from_pandas(tr[['anchor','target','context','title','score']])
    ds_va = HFDataset.from_pandas(va[['anchor','target','context','title','score']])
    ds_te = HFDataset.from_pandas(test_df[['anchor','target','context','title']])

    ds_tr = ds_tr.map(prepare_features, batched=True, remove_columns=ds_tr.column_names)
    ds_va = ds_va.map(prepare_features, batched=True, remove_columns=ds_va.column_names)
    ds_te = ds_te.map(prepare_test_features, batched=True, remove_columns=ds_te.column_names)

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=1, problem_type='regression')
    if n_added:
        model.resize_token_embeddings(len(tokenizer))

    args = TrainingArguments(
        output_dir=f'./fold_base_{fold}',
        num_train_epochs=3,
        per_device_train_batch_size=4,
        per_device_eval_batch_size=8,
        gradient_accumulation_steps=4,
        learning_rate=3e-5,
        weight_decay=0.01,
        warmup_ratio=0.15,
        lr_scheduler_type='cosine',
        logging_steps=50,
        save_strategy='epoch',
        evaluation_strategy='epoch',
        load_best_model_at_end=True,
        metric_for_best_model='pearson',
        greater_is_better=True,
        save_total_limit=1,
        fp16=False,
        dataloader_num_workers=0,
        report_to=None,
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=ds_tr,
        eval_dataset=ds_va,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=1, early_stopping_threshold=2e-4)],
    )
    trainer.train()

    def predict_tta(trainer, dataset, df, is_test=False):
        normal = trainer.predict(dataset).predictions.flatten()
        swapped_df = df.copy()
        swapped_df[['anchor','target']] = swapped_df[['target','anchor']]
        swapped = HFDataset.from_pandas(swapped_df[['anchor','target','context','title']] if is_test else swapped_df[['anchor','target','context','title','score']])
        swapped = swapped.map(prepare_test_features if is_test else prepare_features, batched=True, remove_columns=swapped.column_names)
        swapped_pred = trainer.predict(swapped).predictions.flatten()
        return np.clip((normal + swapped_pred)/2, 0, 1)

    va_pred = predict_tta(trainer, ds_va, va, is_test=False)
    oof_base[va_idx] = va_pred
    fold_pearson = pearsonr(va_pred, va['score'].values)[0]
    fold_pearsons_base.append(fold_pearson)
    print(f'Fold {fold+1} Pearson: {fold_pearson:.4f}')

    test_preds_base[fold] = predict_tta(trainer, ds_te, test_df, is_test=True)

importlib.metadata.version = original_version

oof_pearson_base = pearsonr(oof_base, train_df['score'].values)[0]
print(f'\nFull OOF Pearson (base): {oof_pearson_base:.4f}')
print(f'Mean fold Pearson (base): {np.mean(fold_pearsons_base):.4f} +/- {np.std(fold_pearsons_base):.4f}')

iso_base = IsotonicRegression(out_of_bounds='clip')
iso_base.fit(oof_base, train_df['score'].values)
oof_cal_base = iso_base.predict(oof_base)
oof_cal_pearson_base = pearsonr(oof_cal_base, train_df['score'].values)[0]
print(f'Calibrated OOF Pearson (base): {oof_cal_pearson_base:.4f}')

test_avg_base = np.mean(test_preds_base, axis=0)
test_final_base = np.clip(iso_base.predict(test_avg_base) if oof_cal_pearson_base > oof_pearson_base else test_avg_base, 0, 1)

np.save('oof_base.npy', oof_base)
np.save('test_base.npy', test_avg_base)
print('Saved base model OOF and test predictions to .npy files')
```

```python
import pandas as pd, numpy as np
from scipy.stats import pearsonr
from sklearn.isotonic import IsotonicRegression

# Required assets
oof_large = np.load('oof_large.npy')          # partial OOF (NaNs)
test_large = np.load('test_large.npy')        # full test
oof_base  = np.load('oof_base.npy')           # full OOF
test_base = np.load('test_base.npy')          # full test

# Optional tiny splash (if exists)
try:
    oof_base_pt = np.load('oof_base_pt_partial.npy')
    test_base_pt = np.load('test_base_pt_partial.npy')
    has_base_pt = True
except Exception:
    has_base_pt = False

y = pd.read_csv('train.csv')['score'].values
test_id = pd.read_csv('test.csv')['id']

def nan_aware_blend(arrs, ws):
    M = np.column_stack(arrs).astype(float)
    avail = ~np.isnan(M)
    M = np.nan_to_num(M, nan=0.0)
    w = np.asarray(ws, dtype=float)
    num = (M * w).sum(axis=1)
    den = (avail * w).sum(axis=1)
    den = np.clip(den, 1e-12, None)
    return num / den

# Weights: aggressively favor large; small stabilizer from base; tiny base_pt if present
if has_base_pt:
    w = [0.85, 0.14, 0.01]
    oof_blend = nan_aware_blend([oof_large, oof_base, oof_base_pt], w)
    test_blend = w[0]*test_large + w[1]*test_base + w[2]*test_base_pt
else:
    w = [0.85, 0.15]
    oof_blend = nan_aware_blend([oof_large, oof_base], w)
    test_blend = w[0]*test_large + w[1]*test_base

# Calibrate on full blended OOF
p_uncal = pearsonr(oof_blend, y)[0]
iso = IsotonicRegression(out_of_bounds='clip', increasing=True)
iso.fit(oof_blend, y)
oof_cal = iso.predict(oof_blend)
p_cal = pearsonr(oof_cal, y)[0]
use_cal = p_cal >= p_uncal

# Predict test
test_final = np.clip(iso.predict(test_blend), 0, 1) if use_cal else np.clip(test_blend, 0, 1)
print(f'OOF Pearson: raw {p_uncal:.5f} | cal {p_cal:.5f} | using {"cal" if use_cal else "raw"}')

pd.DataFrame({'id': test_id, 'score': test_final}).to_csv('submission.csv', index=False)
print('submission.csv saved')
```

```python
import pandas as pd, numpy as np
from sklearn.model_selection import GroupKFold
from scipy.stats import pearsonr
from sklearn.isotonic import IsotonicRegression
from sklearn.feature_extraction.text import TfidfVectorizer
import torch, importlib.metadata
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding, EarlyStoppingCallback
from datasets import Dataset as HFDataset
import random, os
random.seed(42); np.random.seed(42); torch.manual_seed(42)

# Small model re-run with in-fold TF-IDF pseudo-titles (contingency if ensemble <0.86)
model_name = 'microsoft/deberta-v3-small'
tokenizer = AutoTokenizer.from_pretrained(model_name)

train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')

# Fallback titles (class_map as before)
sec_map = {
    'A': 'Human Necessities, Medical, Agriculture',
    'B': 'Operations, Transport, Separation, Mixing',
    'C': 'Chemistry, Metallurgy, Combinatorial Tech',
    'D': 'Textiles, Paper, Building Materials',
    'E': 'Fixed Constructions, Building, Earth Moving',
    'F': 'Mechanical Engineering, Lighting, Heating',
    'G': 'Physics, Computing, Calculating, Counting',
    'H': 'Electricity, Basic Electric Elements',
    'Y': 'Emerging Technologies, Cross-Sectional'
}
class_map = {
    'A61': 'Medical or Veterinary Science; Hygiene',
    'C07': 'Organic Chemistry',
    'C08': 'Organic Macromolecular Compounds; Preparation or Chemical Working-Up',
    'C12': 'Biochemistry; Beer; Spirits; Wine; Vinegar; Microbiology; Enzymology; Mutation or Genetic Engineering',
    'G01': 'Measuring; Testing',
    'G06': 'Computing; Calculating; Counting',
    'H01': 'Basic Electric Elements',
    'H04': 'Electric Communication Technique',
    'B60': 'Vehicles in General',
    'F16': 'Engineering Elements or Units; General Measures for Producing and Maintaining Effective Functioning of Machines or Installations',
    'H05': 'Electric Techniques Not Otherwise Provided For',
    'H03': 'Basic Electronic Circuitry',
    'H01L': 'Semiconductor Devices; Electric Solid State Devices Not Otherwise Provided For',
    'A61K': 'Preparations for Medical, Dental, or Toilet Purposes'
}
def get_title(context):
    cls = context[:3] if len(context) >= 3 else context[0]
    sec = context[0]
    return class_map.get(cls, sec_map.get(sec, 'no title'))
train_df['title'] = train_df['context'].apply(get_title)
test_df['title']  = test_df['context'].apply(get_title)

# Build hierarchical tokens (same as base)
all_ctx = pd.concat([train_df['context'], test_df['context']]).unique()
def get_hier_parts(code):
    sec = code[0] if code else 'X'
    cls = code[:3] if len(code) >= 3 else sec
    sub = code[:4] if len(code) >= 4 else cls
    grp = code[:7].replace('/', '') if len(code) >= 7 else sub
    return f"<SEC_{sec}>", f"<CLS_{cls}>", f"<SUB_{sub}>", f"<GRP_{grp}>"
special_tokens = set()
for c in all_ctx:
    special_tokens.update(get_hier_parts(c))
n_added = tokenizer.add_special_tokens({'additional_special_tokens': sorted(special_tokens)})
print(f"Added {n_added} hierarchical tokens. Vocab: {len(tokenizer)}")
def hier_str(code):
    return " ".join(get_hier_parts(code)) + " "

# Build full train pseudo-titles for test (outside loop to avoid leakage)
def build_pseudo_titles(df, top_k=30, ngram_range=(1,3)):
    tmp = df.copy()
    tmp['text'] = (tmp['anchor'].astype(str) + ' ' + tmp['target'].astype(str)).str.lower()
    grp = tmp.groupby('context')['text'].apply(lambda x: ' ||| '.join(x))
    contexts, corpus = grp.index.tolist(), grp.tolist()
    vec = TfidfVectorizer(ngram_range=ngram_range, min_df=2, max_features=30000, stop_words=None, norm='l2')
    X = vec.fit_transform(corpus)
    vocab = np.array(vec.get_feature_names_out())
    titles = {}
    for i, ctx in enumerate(contexts):
        row = X.getrow(i)
        if row.nnz == 0: 
            continue
        data = row.toarray().ravel()
        inds = np.argsort(-data)[:top_k]
        titles[ctx] = ' '.join(vocab[inds])
    return titles

title_map_full = build_pseudo_titles(train_df, top_k=30, ngram_range=(1,3))

def prepare_features(examples):
    h = [hier_str(c) for c in examples['context']]
    pt = [title_map_full.get(c, '') for c in examples['context']]
    t1 = [f"{hh}{p} anchor: {a} [CPC] {c}" for hh,p,a,c in zip(h, pt, examples['anchor'], examples['context'])]
    t2 = [f"{hh}{p} target: {b} [CPC] {c}" for hh,p,b,c in zip(h, pt, examples['target'], examples['context'])]
    mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    if 'score' in examples:
        mi['labels'] = examples['score']
    return mi

def prepare_test_features(examples):
    h = [hier_str(c) for c in examples['context']]
    pt = [title_map_full.get(c, '') for c in examples['context']]
    t1 = [f"{hh}{p} anchor: {a} [CPC] {c}" for hh,p,a,c in zip(h, pt, examples['anchor'], examples['context'])]
    t2 = [f"{hh}{p} target: {b} [CPC] {c}" for hh,p,b,c in zip(h, pt, examples['target'], examples['context'])]
    mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    return mi

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    preds = np.clip(preds.flatten(), 0, 1)
    return {'pearson': pearsonr(preds, labels)[0]}

gkf = GroupKFold(n_splits=3)
oof_small_tfidf = np.zeros(len(train_df))
test_preds_small_tfidf = np.zeros((3, len(test_df)))
fold_pearsons_small_tfidf = []

# Patch
original_version = importlib.metadata.version
def patched_version(name):
    if name == 'torch': return '2.4.1'
    return original_version(name)
importlib.metadata.version = patched_version

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

for fold, (tr_idx, va_idx) in enumerate(gkf.split(train_df, groups=train_df['anchor'])):
    print(f'\n--- Small TF-IDF Fold {fold+1}/3 ---')
    tr = train_df.iloc[tr_idx]
    va = train_df.iloc[va_idx]

    # Build pseudo-titles from tr only for this fold (tr/va)
    title_map_fold = build_pseudo_titles(tr, top_k=30, ngram_range=(1,3))
    tr = tr.copy(); va = va.copy(); te = test_df.copy()
    tr['title'] = tr['context'].map(title_map_fold).fillna(tr['title'])
    va['title'] = va['context'].map(title_map_fold).fillna(va['title'])
    te['title'] = te['context'].map(title_map_full).fillna(te['title'])  # Use full train map for test

    ds_tr = HFDataset.from_pandas(tr[['anchor','target','context','title','score']])
    ds_va = HFDataset.from_pandas(va[['anchor','target','context','title','score']])
    ds_te = HFDataset.from_pandas(te[['anchor','target','context','title']])

    ds_tr = ds_tr.map(prepare_features, batched=True, remove_columns=ds_tr.column_names)
    ds_va = ds_va.map(prepare_features, batched=True, remove_columns=ds_va.column_names)
    ds_te = ds_te.map(prepare_test_features, batched=True, remove_columns=ds_te.column_names)

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=1, problem_type='regression')
    if n_added:
        model.resize_token_embeddings(len(tokenizer))

    args = TrainingArguments(
        output_dir=f'./fold_small_tfidf_{fold}',
        num_train_epochs=2,  # Quick
        per_device_train_batch_size=4,
        per_device_eval_batch_size=8,
        gradient_accumulation_steps=4,
        learning_rate=3e-5,
        weight_decay=0.01,
        warmup_ratio=0.15,
        lr_scheduler_type='cosine',
        logging_steps=50,
        save_strategy='epoch',
        evaluation_strategy='epoch',
        load_best_model_at_end=True,
        metric_for_best_model='pearson',
        greater_is_better=True,
        save_total_limit=1,
        fp16=False,
        dataloader_num_workers=0,
        report_to=None,
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=ds_tr,
        eval_dataset=ds_va,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=1, early_stopping_threshold=2e-4)],
    )
    trainer.train()

    def predict_tta(trainer, dataset, df, is_test=False):
        normal = trainer.predict(dataset).predictions.flatten()
        swapped_df = df.copy()
        swapped_df[['anchor','target']] = swapped_df[['target','anchor']]
        swapped = HFDataset.from_pandas(swapped_df[['anchor','target','context','title']] if is_test else swapped_df[['anchor','target','context','title','score']])
        swapped = swapped.map(prepare_test_features if is_test else prepare_features, batched=True, remove_columns=swapped.column_names)
        swapped_pred = trainer.predict(swapped).predictions.flatten()
        return np.clip((normal + swapped_pred)/2, 0, 1)

    va_pred = predict_tta(trainer, ds_va, va, is_test=False)
    oof_small_tfidf[va_idx] = va_pred
    fold_pearson = pearsonr(va_pred, va['score'].values)[0]
    fold_pearsons_small_tfidf.append(fold_pearson)
    print(f'Small TF-IDF Fold {fold+1} Pearson: {fold_pearson:.4f}')

    test_preds_small_tfidf[fold] = predict_tta(trainer, ds_te, te, is_test=True)

importlib.metadata.version = original_version

oof_pearson_small_tfidf = pearsonr(oof_small_tfidf, train_df['score'].values)[0]
print(f'\nFull OOF Pearson (small TF-IDF): {oof_pearson_small_tfidf:.4f}')
print(f'Mean fold Pearson (small TF-IDF): {np.mean(fold_pearsons_small_tfidf):.4f} +/- {np.std(fold_pearsons_small_tfidf):.4f}')

iso_small_tfidf = IsotonicRegression(out_of_bounds='clip')
iso_small_tfidf.fit(oof_small_tfidf, train_df['score'].values)
oof_cal_small_tfidf = iso_small_tfidf.predict(oof_small_tfidf)
oof_cal_pearson_small_tfidf = pearsonr(oof_cal_small_tfidf, train_df['score'].values)[0]
print(f'Calibrated OOF Pearson (small TF-IDF): {oof_cal_pearson_small_tfidf:.4f}')

test_avg_small_tfidf = np.mean(test_preds_small_tfidf, axis=0)
test_final_small_tfidf = np.clip(iso_small_tfidf.predict(test_avg_small_tfidf) if oof_cal_pearson_small_tfidf > oof_pearson_small_tfidf else test_avg_small_tfidf, 0, 1)

np.save('oof_small_tfidf.npy', oof_small_tfidf)
np.save('test_small_tfidf.npy', test_avg_small_tfidf)
print('Saved small TF-IDF model OOF and test predictions to .npy files')
```

```python
import pandas as pd, numpy as np
from sklearn.model_selection import GroupKFold
from scipy.stats import pearsonr
from sklearn.isotonic import IsotonicRegression
import torch, importlib.metadata
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding, EarlyStoppingCallback
from datasets import Dataset as HFDataset
import random, os
random.seed(42); np.random.seed(42); torch.manual_seed(42)

# Load data
train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')

# Fallback titles (class_map as before)
sec_map = {
    'A': 'Human Necessities, Medical, Agriculture',
    'B': 'Operations, Transport, Separation, Mixing',
    'C': 'Chemistry, Metallurgy, Combinatorial Tech',
    'D': 'Textiles, Paper, Building Materials',
    'E': 'Fixed Constructions, Building, Earth Moving',
    'F': 'Mechanical Engineering, Lighting, Heating',
    'G': 'Physics, Computing, Calculating, Counting',
    'H': 'Electricity, Basic Electric Elements',
    'Y': 'Emerging Technologies, Cross-Sectional'
}
class_map = {
    'A61': 'Medical or Veterinary Science; Hygiene',
    'C07': 'Organic Chemistry',
    'C08': 'Organic Macromolecular Compounds; Preparation or Chemical Working-Up',
    'C12': 'Biochemistry; Beer; Spirits; Wine; Vinegar; Microbiology; Enzymology; Mutation or Genetic Engineering',
    'G01': 'Measuring; Testing',
    'G06': 'Computing; Calculating; Counting',
    'H01': 'Basic Electric Elements',
    'H04': 'Electric Communication Technique',
    'B60': 'Vehicles in General',
    'F16': 'Engineering Elements or Units; General Measures for Producing and Maintaining Effective Functioning of Machines or Installations',
    'H05': 'Electric Techniques Not Otherwise Provided For',
    'H03': 'Basic Electronic Circuitry',
    'H01L': 'Semiconductor Devices; Electric Solid State Devices Not Otherwise Provided For',
    'A61K': 'Preparations for Medical, Dental, or Toilet Purposes'
}
def get_title(context):
    cls = context[:3] if len(context) >= 3 else context[0]
    sec = context[0]
    return class_map.get(cls, sec_map.get(sec, 'no title'))
train_df['title'] = train_df['context'].apply(get_title)
test_df['title']  = test_df['context'].apply(get_title)

# Build hierarchical CPC tokens from train+test
all_ctx = pd.concat([train_df['context'], test_df['context']]).unique()

def get_hier_parts(code):
    sec = code[0] if code else 'X'
    cls = code[:3] if len(code) >= 3 else sec
    sub = code[:4] if len(code) >= 4 else cls
    grp = code[:7].replace('/', '') if len(code) >= 7 else sub  # e.g., A61K31
    return f"<SEC_{sec}>", f"<CLS_{cls}>", f"<SUB_{sub}>", f"<GRP_{grp}>"

special_tokens = set()
for c in all_ctx:
    special_tokens.update(get_hier_parts(c))

model_name = 'microsoft/deberta-v3-base'
tokenizer = AutoTokenizer.from_pretrained(model_name)
n_added = tokenizer.add_special_tokens({'additional_special_tokens': sorted(special_tokens)})
print(f"Added {n_added} hierarchical tokens. Vocab: {len(tokenizer)}")

def hier_str(code):
    return " ".join(get_hier_parts(code)) + " "

# Build full train pseudo-titles for test (outside loop to avoid leakage)
def build_pseudo_titles(df, top_k=30, ngram_range=(1,3)):
    tmp = df.copy()
    tmp['text'] = (tmp['anchor'].astype(str) + ' ' + tmp['target'].astype(str)).str.lower()
    grp = tmp.groupby('context')['text'].apply(lambda x: ' ||| '.join(x))
    contexts, corpus = grp.index.tolist(), grp.tolist()
    vec = TfidfVectorizer(ngram_range=ngram_range, min_df=2, max_features=30000, stop_words=None, norm='l2')
    X = vec.fit_transform(corpus)
    vocab = np.array(vec.get_feature_names_out())
    titles = {}
    for i, ctx in enumerate(contexts):
        row = X.getrow(i)
        if row.nnz == 0: 
            continue
        data = row.toarray().ravel()
        inds = np.argsort(-data)[:top_k]
        titles[ctx] = ' '.join(vocab[inds])
    return titles

title_map_full = build_pseudo_titles(train_df, top_k=30, ngram_range=(1,3))

def prepare_features(examples):
    h = [hier_str(c) for c in examples['context']]
    pt = [title_map_full.get(c, '') for c in examples['context']]
    t1 = [f"{hh}{p} anchor: {a} [CPC] {c}" for hh,p,a,c in zip(h, pt, examples['anchor'], examples['context'])]
    t2 = [f"{hh}{p} target: {b} [CPC] {c}" for hh,p,b,c in zip(h, pt, examples['target'], examples['context'])]
    mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    if 'score' in examples:
        mi['labels'] = examples['score']
    return mi

def prepare_test_features(examples):
    h = [hier_str(c) for c in examples['context']]
    pt = [title_map_full.get(c, '') for c in examples['context']]
    t1 = [f"{hh}{p} anchor: {a} [CPC] {c}" for hh,p,a,c in zip(h, pt, examples['anchor'], examples['context'])]
    t2 = [f"{hh}{p} target: {b} [CPC] {c}" for hh,p,b,c in zip(h, pt, examples['target'], examples['context'])]
    mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    return mi

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    preds = np.clip(preds.flatten(), 0, 1)
    return {'pearson': pearsonr(preds, labels)[0]}

gkf = GroupKFold(n_splits=3)
oof_base_pt = np.zeros(len(train_df))
test_preds_base_pt = np.zeros((3, len(test_df)))
fold_pearsons_base_pt = []

# Patch torch metadata for +cpu wheel
original_version = importlib.metadata.version
def patched_version(name):
    if name == 'torch': return '2.4.1'
    return original_version(name)
importlib.metadata.version = patched_version

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

for fold, (tr_idx, va_idx) in enumerate(gkf.split(train_df, groups=train_df['anchor'])):
    print(f'\n--- Base PT Fold {fold+1}/3 ---')
    tr = train_df.iloc[tr_idx]
    va = train_df.iloc[va_idx]

    # Build pseudo-titles from tr only for this fold (tr/va)
    title_map_fold = build_pseudo_titles(tr, top_k=30, ngram_range=(1,3))
    tr = tr.copy(); va = va.copy(); te = test_df.copy()
    tr['title'] = tr['context'].map(title_map_fold).fillna(tr['title'])
    va['title'] = va['context'].map(title_map_fold).fillna(va['title'])
    te['title'] = te['context'].map(title_map_full).fillna(te['title'])  # Use full train map for test

    ds_tr = HFDataset.from_pandas(tr[['anchor','target','context','title','score']])
    ds_va = HFDataset.from_pandas(va[['anchor','target','context','title','score']])
    ds_te = HFDataset.from_pandas(te[['anchor','target','context','title']])

    ds_tr = ds_tr.map(prepare_features, batched=True, remove_columns=ds_tr.column_names)
    ds_va = ds_va.map(prepare_features, batched=True, remove_columns=ds_va.column_names)
    ds_te = ds_te.map(prepare_test_features, batched=True, remove_columns=ds_te.column_names)

    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=1, problem_type='regression')
    if n_added:
        model.resize_token_embeddings(len(tokenizer))

    args = TrainingArguments(
        output_dir=f'./fold_base_pt_{fold}',
        num_train_epochs=3,
        per_device_train_batch_size=4,
        per_device_eval_batch_size=8,
        gradient_accumulation_steps=4,
        learning_rate=2e-5,
        weight_decay=0.01,
        warmup_ratio=0.15,
        lr_scheduler_type='cosine',
        logging_steps=50,
        save_strategy='epoch',
        evaluation_strategy='epoch',
        load_best_model_at_end=True,
        metric_for_best_model='pearson',
        greater_is_better=True,
        save_total_limit=1,
        fp16=False,
        dataloader_num_workers=0,
        report_to=None,
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=ds_tr,
        eval_dataset=ds_va,
        tokenizer=tokenizer,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=1, early_stopping_threshold=2e-4)],
    )
    trainer.train()

    def predict_tta(trainer, dataset, df, is_test=False):
        normal = trainer.predict(dataset).predictions.flatten()
        swapped_df = df.copy()
        swapped_df[['anchor','target']] = swapped_df[['target','anchor']]
        swapped = HFDataset.from_pandas(swapped_df[['anchor','target','context','title']] if is_test else swapped_df[['anchor','target','context','title','score']])
        swapped = swapped.map(prepare_test_features if is_test else prepare_features, batched=True, remove_columns=swapped.column_names)
        swapped_pred = trainer.predict(swapped).predictions.flatten()
        return np.clip((normal + swapped_pred)/2, 0, 1)

    va_pred = predict_tta(trainer, ds_va, va, is_test=False)
    oof_base_pt[va_idx] = va_pred
    fold_pearson = pearsonr(va_pred, va['score'].values)[0]
    fold_pearsons_base_pt.append(fold_pearson)
    print(f'Base PT Fold {fold+1} Pearson: {fold_pearson:.4f}')

    test_preds_base_pt[fold] = predict_tta(trainer, ds_te, te, is_test=True)

importlib.metadata.version = original_version

oof_pearson_base_pt = pearsonr(oof_base_pt, train_df['score'].values)[0]
print(f'\nFull OOF Pearson (base_pt): {oof_pearson_base_pt:.4f}')
print(f'Mean fold Pearson (base_pt): {np.mean(fold_pearsons_base_pt):.4f} +/- {np.std(fold_pearsons_base_pt):.4f}')

iso_base_pt = IsotonicRegression(out_of_bounds='clip')
iso_base_pt.fit(oof_base_pt, train_df['score'].values)
oof_cal_base_pt = iso_base_pt.predict(oof_base_pt)
oof_cal_pearson_base_pt = pearsonr(oof_cal_base_pt, train_df['score'].values)[0]
print(f'Calibrated OOF Pearson (base_pt): {oof_cal_pearson_base_pt:.4f}')

test_avg_base_pt = np.mean(test_preds_base_pt, axis=0)
test_final_base_pt = np.clip(iso_base_pt.predict(test_avg_base_pt) if oof_cal_pearson_base_pt > oof_pearson_base_pt else test_avg_base_pt, 0, 1)

np.save('oof_base_pt.npy', oof_base_pt)
np.save('test_base_pt.npy', test_avg_base_pt)
print('Saved base_pt model OOF and test predictions to .npy files')
```

```python
import pandas as pd, numpy as np
from sklearn.model_selection import GroupKFold
from scipy.stats import pearsonr
from sklearn.isotonic import IsotonicRegression
import torch, importlib.metadata
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, DataCollatorWithPadding
from datasets import Dataset as HFDataset
from sklearn.feature_extraction.text import TfidfVectorizer
import random, os
random.seed(42); np.random.seed(42); torch.manual_seed(42)

# Load data and setup (same as Cell 15)
train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')

# Fallback titles
sec_map = {
    'A': 'Human Necessities, Medical, Agriculture',
    'B': 'Operations, Transport, Separation, Mixing',
    'C': 'Chemistry, Metallurgy, Combinatorial Tech',
    'D': 'Textiles, Paper, Building Materials',
    'E': 'Fixed Constructions, Building, Earth Moving',
    'F': 'Mechanical Engineering, Lighting, Heating',
    'G': 'Physics, Computing, Calculating, Counting',
    'H': 'Electricity, Basic Electric Elements',
    'Y': 'Emerging Technologies, Cross-Sectional'
}
class_map = {
    'A61': 'Medical or Veterinary Science; Hygiene',
    'C07': 'Organic Chemistry',
    'C08': 'Organic Macromolecular Compounds; Preparation or Chemical Working-Up',
    'C12': 'Biochemistry; Beer; Spirits; Wine; Vinegar; Microbiology; Enzymology; Mutation or Genetic Engineering',
    'G01': 'Measuring; Testing',
    'G06': 'Computing; Calculating; Counting',
    'H01': 'Basic Electric Elements',
    'H04': 'Electric Communication Technique',
    'B60': 'Vehicles in General',
    'F16': 'Engineering Elements or Units; General Measures for Producing and Maintaining Effective Functioning of Machines or Installations',
    'H05': 'Electric Techniques Not Otherwise Provided For',
    'H03': 'Basic Electronic Circuitry',
    'H01L': 'Semiconductor Devices; Electric Solid State Devices Not Otherwise Provided For',
    'A61K': 'Preparations for Medical, Dental, or Toilet Purposes'
}
def get_title(context):
    cls = context[:3] if len(context) >= 3 else context[0]
    sec = context[0]
    return class_map.get(cls, sec_map.get(sec, 'no title'))
train_df['title'] = train_df['context'].apply(get_title)
test_df['title']  = test_df['context'].apply(get_title)

# Hierarchical tokens
all_ctx = pd.concat([train_df['context'], test_df['context']]).unique()
def get_hier_parts(code):
    sec = code[0] if code else 'X'
    cls = code[:3] if len(code) >= 3 else sec
    sub = code[:4] if len(code) >= 4 else cls
    grp = code[:7].replace('/', '') if len(code) >= 7 else sub
    return f"<SEC_{sec}>", f"<CLS_{cls}>", f"<SUB_{sub}>", f"<GRP_{grp}>"
special_tokens = set()
for c in all_ctx:
    special_tokens.update(get_hier_parts(c))

model_name = 'microsoft/deberta-v3-base'
tokenizer = AutoTokenizer.from_pretrained(model_name)
n_added = tokenizer.add_special_tokens({'additional_special_tokens': sorted(special_tokens)})
print(f"Added {n_added} hierarchical tokens. Vocab: {len(tokenizer)}")

def hier_str(code):
    return " ".join(get_hier_parts(code)) + " "

# Full train pseudo-titles for test
def build_pseudo_titles(df, top_k=30, ngram_range=(1,3)):
    tmp = df.copy()
    tmp['text'] = (tmp['anchor'].astype(str) + ' ' + tmp['target'].astype(str)).str.lower()
    grp = tmp.groupby('context')['text'].apply(lambda x: ' ||| '.join(x))
    contexts, corpus = grp.index.tolist(), grp.tolist()
    vec = TfidfVectorizer(ngram_range=ngram_range, min_df=2, max_features=30000, stop_words=None, norm='l2')
    X = vec.fit_transform(corpus)
    vocab = np.array(vec.get_feature_names_out())
    titles = {}
    for i, ctx in enumerate(contexts):
        row = X.getrow(i)
        if row.nnz == 0: 
            continue
        data = row.toarray().ravel()
        inds = np.argsort(-data)[:top_k]
        titles[ctx] = ' '.join(vocab[inds])
    return titles

title_map_full = build_pseudo_titles(train_df, top_k=30, ngram_range=(1,3))

def prepare_features(examples):
    h = [hier_str(c) for c in examples['context']]
    pt = examples['title']  # Use passed title (fold-specific for val)
    t1 = [f"{hh}{p} anchor: {a} [CPC] {c}" for hh,p,a,c in zip(h, pt, examples['anchor'], examples['context'])]
    t2 = [f"{hh}{p} target: {b} [CPC] {c}" for hh,p,b,c in zip(h, pt, examples['target'], examples['context'])]
    mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    if 'score' in examples:
        mi['labels'] = examples['score']
    return mi

def prepare_test_features(examples):
    h = [hier_str(c) for c in examples['context']]
    pt = [title_map_full.get(c, '') for c in examples['context']]  # Full for test
    t1 = [f"{hh}{p} anchor: {a} [CPC] {c}" for hh,p,a,c in zip(h, pt, examples['anchor'], examples['context'])]
    t2 = [f"{hh}{p} target: {b} [CPC] {c}" for hh,p,b,c in zip(h, pt, examples['target'], examples['context'])]
    mi = tokenizer(t1, t2, max_length=160, truncation=True, padding=False)
    return mi

gkf = GroupKFold(n_splits=3)
oof_base_pt_partial = np.full(len(train_df), np.nan)
test_preds_base_pt_partial = np.zeros((2, len(test_df)))  # Only 2 folds
fold_pearsons_base_pt_partial = []

# Patch
original_version = importlib.metadata.version
def patched_version(name):
    if name == 'torch': return '2.4.1'
    return original_version(name)
importlib.metadata.version = patched_version

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Available folds: 0 and 1 (Folds 1/2 in 0-index)
available_folds = [0, 1]

for fold in available_folds:
    print(f'\n--- Inference on saved Fold {fold} (no leak) ---')
    tr_idx, va_idx = list(gkf.split(train_df, groups=train_df['anchor']))[fold]
    tr = train_df.iloc[tr_idx]
    va = train_df.iloc[va_idx]

    # Build fold-specific pseudo-titles from tr only
    title_map_fold = build_pseudo_titles(tr, top_k=30, ngram_range=(1,3))
    tr = tr.copy(); va = va.copy(); te = test_df.copy()
    tr['title'] = tr['context'].map(title_map_fold).fillna(tr['title'])
    va['title'] = va['context'].map(title_map_fold).fillna(va['title'])  # In-fold for val (no leak)
    te['title'] = te['context'].map(title_map_full).fillna(te['title'])  # Full for test

    ds_va = HFDataset.from_pandas(va[['anchor','target','context','title','score']])
    ds_te = HFDataset.from_pandas(te[['anchor','target','context','title']])

    ds_va = ds_va.map(prepare_features, batched=True, remove_columns=ds_va.column_names)
    ds_te = ds_te.map(prepare_test_features, batched=True, remove_columns=ds_te.column_names)

    # Load saved model from checkpoint
    model_dir = f'./fold_base_pt_{fold}/checkpoint-4101'
    model = AutoModelForSequenceClassification.from_pretrained(model_dir, num_labels=1, problem_type='regression')
    if n_added:
        model.resize_token_embeddings(len(tokenizer))
    model.eval()

    # Dummy trainer for predict (no training args needed)
    dummy_args = TrainingArguments(output_dir='./dummy', per_device_eval_batch_size=8, fp16=False)
    trainer = Trainer(
        model=model,
        args=dummy_args,
        tokenizer=tokenizer,
        data_collator=data_collator,
    )

    def predict_tta(trainer, dataset, df, is_test=False):
        normal = trainer.predict(dataset).predictions.flatten()
        swapped_df = df.copy()
        swapped_df[['anchor','target']] = swapped_df[['target','anchor']]
        swapped = HFDataset.from_pandas(swapped_df[['anchor','target','context','title']] if is_test else swapped_df[['anchor','target','context','title','score']])
        swapped = swapped.map(prepare_test_features if is_test else prepare_features, batched=True, remove_columns=swapped.column_names)
        swapped_pred = trainer.predict(swapped).predictions.flatten()
        return np.clip((normal + swapped_pred)/2, 0, 1)

    va_pred = predict_tta(trainer, ds_va, va, is_test=False)
    oof_base_pt_partial[va_idx] = va_pred
    fold_pearson = pearsonr(va_pred, va['score'].values)[0]
    fold_pearsons_base_pt_partial.append(fold_pearson)
    print(f'Fold {fold} Partial Pearson: {fold_pearson:.4f}')

    test_preds_base_pt_partial[fold] = predict_tta(trainer, ds_te, te, is_test=True)

importlib.metadata.version = original_version

# Average available test preds (2 folds)
test_avg_base_pt_partial = np.mean(test_preds_base_pt_partial, axis=0)

# Partial OOF stats (only on covered rows)
valid_mask = ~np.isnan(oof_base_pt_partial)
partial_oof_pearson = pearsonr(oof_base_pt_partial[valid_mask], train_df['score'][valid_mask].values)[0]
print(f'\nPartial OOF Pearson (2 folds): {partial_oof_pearson:.4f}')
print(f'Mean fold Pearson (partial): {np.mean(fold_pearsons_base_pt_partial):.4f} +/- {np.std(fold_pearsons_base_pt_partial):.4f}')

# Calibrate on partial OOF
iso_base_pt_partial = IsotonicRegression(out_of_bounds='clip')
iso_base_pt_partial.fit(oof_base_pt_partial[valid_mask], train_df['score'][valid_mask].values)
oof_cal_partial = iso_base_pt_partial.predict(oof_base_pt_partial[valid_mask])
oof_cal_pearson_partial = pearsonr(oof_cal_partial, train_df['score'][valid_mask].values)[0]
print(f'Calibrated Partial OOF Pearson: {oof_cal_pearson_partial:.4f}')

# For ensemble: save partial oof (NaN for fold 2) and averaged test
np.save('oof_base_pt_partial.npy', oof_base_pt_partial)
np.save('test_base_pt_partial.npy', test_avg_base_pt_partial)
print('Saved partial base_pt OOF (with NaNs) and averaged test predictions to .npy files')
```
